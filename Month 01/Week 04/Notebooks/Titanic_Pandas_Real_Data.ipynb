## 1. 🚢 About the Titanic Dataset

The **RMS Titanic** sank on 15 April 1912 after hitting an iceberg.  
Of the **2,224 passengers and crew**, only **710 survived**.  

This dataset contains information about **891 passengers** and whether they survived.

### Column Dictionary

| Column | Description | Type |
|--------|-------------|------|
| `PassengerId` | Unique ID for each passenger | Integer |
| `Survived` | 0 = Died, 1 = Survived | Integer (Target) |
| `Pclass` | Ticket class: 1 = 1st, 2 = 2nd, 3 = 3rd | Integer |
| `Name` | Full name of passenger | String |
| `Sex` | male / female | String |
| `Age` | Age in years | Float (has missing) |
| `SibSp` | # of siblings/spouses aboard | Integer |
| `Parch` | # of parents/children aboard | Integer |
| `Ticket` | Ticket number | String |
| `Fare` | Passenger fare (price paid) | Float |
| `Cabin` | Cabin number | String (mostly missing) |
| `Embarked` | Port of Embarkation: C=Cherbourg, Q=Queenstown, S=Southampton | String |

> 📥 **Download:** Go to `kaggle.com/c/titanic` → Data tab → Download `train.csv`

## 8. 🛠️ Step 6 — Add New Features (Feature Engineering)

Creating new columns from existing ones — this is where Pandas skills directly power ML.

### 8.1 Extract Title from Name

````python
# Passenger names follow this pattern: 'Last, Title. First'
print("Sample names:")
print(df_clean['Name'].head(8).tolist())
````

````python
name = 'Braund,  Mr. Owen Harris'
name.split(',')[1].strip().split('.')[0]
````

````python
# Extract title using apply() + string split
def extract_title(name):
    # Name format: 'Braund, Mr. Owen Harris'
    # Split by comma → ['Braund', ' Mr. Owen Harris']
    # Take part after comma, split by dot → [' Mr', ' Owen Harris']
    # Take first part, strip spaces
    # ['']
    return name.split(',')[1].split('.')[0].strip()

df_clean['Title'] = df_clean['Name'].apply(extract_title)

print("All unique titles found:")
print(df_clean['Title'].value_counts())
````

````python
# Simplify rare titles into groups using map()
# Common: Mr, Miss, Mrs, Master
# Rare: Dr, Rev, Col, Major → group as 'Officer'
# Noble: Don, Lady, Sir, Countess, Jonkheer → group as 'Noble'

def simplify_title(title):
    if title in ['Mr']:                                        return 'Mr'
    elif title in ['Miss', 'Mlle', 'Ms']:                     return 'Miss'
    elif title in ['Mrs', 'Mme']:                             return 'Mrs'
    elif title in ['Master']:                                  return 'Master'
    elif title in ['Dr', 'Rev', 'Col', 'Major', 'Capt']:      return 'Officer'
    else:                                                      return 'Noble'

df_clean['Title'] = df_clean['Title'].apply(simplify_title)

print("Simplified titles:")
print(df_clean['Title'].value_counts())
````

````python
# Survival rate by title — interesting insight!
print("Survival rate by Title:")
title_survival = df_clean.groupby('Title')['Survived'].agg(['mean','count'])
title_survival.columns = ['Survival Rate', 'Count']
title_survival['Survival Rate'] = (title_survival['Survival Rate'] * 100).round(1)
print(title_survival.sort_values('Survival Rate', ascending=False))
````

````python
# Now drop the Name column — we've extracted what we needed
df_clean = df_clean.drop(columns=['Name'])
print("Columns now:", df_clean.columns.tolist())
````

### 8.2 Age Groups

````python
# Create AgeGroup column using apply()
def age_group(age):
    if age < 13:   return 'Child'     # 0-12
    elif age < 18: return 'Teen'      # 13-17
    elif age < 60: return 'Adult'     # 18-59
    else:          return 'Senior'    # 60+

df_clean['AgeGroup'] = df_clean['Age'].apply(age_group)

print("Age group distribution:")
print(df_clean['AgeGroup'].value_counts())
````

````python
# Survival rate by age group
print("Survival rate by Age Group:")
ag_surv = df_clean.groupby('AgeGroup')['Survived'].mean() * 100
print(ag_surv.round(1))
````

### 8.3 Family Size

````python
# SibSp = siblings + spouses
# Parch  = parents + children
# FamilySize = total family members on board (including self)

df_clean['FamilySize'] = df_clean['SibSp'] + df_clean['Parch'] + 1

print("Family size distribution:")
print(df_clean['FamilySize'].value_counts().sort_index())
````

````python
# Is the passenger travelling alone?
df_clean['IsAlone'] = (df_clean['FamilySize'] == 1).astype(int)
# astype(int) converts True→1, False→0

alone_survival = df_clean.groupby('IsAlone')['Survived'].mean() * 100
print("Survival rate:")
print(f"  Travelling with family : {alone_survival[0]:.1f}%")
print(f"  Travelling alone       : {alone_survival[1]:.1f}%")
````

### 8.4 Fare Bands

````python
# Fare distribution
print("Fare statistics:")
print(df_clean['Fare'].describe().round(2))
````

````python
# Create Fare bands using apply()
def fare_band(fare):
    if   fare <= 10:  return 'Low'
    elif fare <= 30:  return 'Medium'
    elif fare <= 100: return 'High'
    else:             return 'Very High'

df_clean['FareBand'] = df_clean['Fare'].apply(fare_band)

print("Fare band distribution:")
print(df_clean['FareBand'].value_counts())
print()
print("Survival rate by Fare Band:")
fare_surv = df_clean.groupby('FareBand')['Survived'].mean() * 100
print(fare_surv.round(1))
````

## 10. Step 8 — Final Model-Ready Data

The last step: encode categorical columns into numbers so an ML model can read them.

````python
# Where are we now?
print("Current DataFrame:")
df_clean.head()
````

````python
print("Current dtypes:")
print(df_clean.dtypes)
print()
print("Remaining categorical (string) columns to encode:")
cat_cols = df_clean.select_dtypes(include='object').columns.tolist()
print(cat_cols)
````

### 10.1 Encode Binary Column — Sex

````python
# Sex: male → 0, female → 1
df_clean['Sex'] = df_clean['Sex'].map({'male': 0, 'female': 1})

print("Sex after encoding:")
print(df_clean['Sex'].value_counts())
````

### 10.2 Encode Multi-Class Columns — Embarked, Title, AgeGroup, FareBand

````python
# Embarked: Southampton→0, Cherbourg→1, Queenstown→2
embarked_map = {'Southampton': 0, 'Cherbourg': 1, 'Queenstown': 2}
df_clean['Embarked'] = df_clean['Embarked'].map(embarked_map)

# Title: encode as numbers
title_map = {'Mr': 0, 'Miss': 1, 'Mrs': 2, 'Master': 3, 'Officer': 4, 'Noble': 5}
df_clean['Title'] = df_clean['Title'].map(title_map)

# AgeGroup: Child→0, Teen→1, Adult→2, Senior→3
age_map = {'Child': 0, 'Teen': 1, 'Adult': 2, 'Senior': 3}
df_clean['AgeGroup'] = df_clean['AgeGroup'].map(age_map)

# FareBand: Low→0, Medium→1, High→2, Very High→3
fare_map_enc = {'Low': 0, 'Medium': 1, 'High': 2, 'Very High': 3}
df_clean['FareBand'] = df_clean['FareBand'].map(fare_map_enc)

print("Encoding done!")
print(df_clean.dtypes)
````

### 10.3 Drop Redundant Columns

````python
# We now have AgeGroup and FareBand which are better than raw Age and Fare
# We also have FamilySize which captures SibSp + Parch together
# So we can drop: Age, Fare, SibSp, Parch

df_clean = df_clean.drop(columns=['SibSp', 'Parch'])

print("Final columns:")
print(df_clean.columns.tolist())
````
