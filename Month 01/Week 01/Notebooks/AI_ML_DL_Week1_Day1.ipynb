## 🤖 Part 1: Understanding Artificial Intelligence

### What is Artificial Intelligence?

**Artificial Intelligence (AI)** is the simulation of human intelligence processes by machines, especially computer systems. These processes include:

- **Learning**: Acquiring information and rules for using it
- **Reasoning**: Using rules to reach approximate or definite conclusions
- **Problem-Solving**: Finding solutions to complex challenges
- **Perception**: Interpreting sensory data (vision, speech)
- **Language Understanding**: Comprehending and generating human language

### AI vs Machine Learning vs Deep Learning

Think of these as nested concepts:

```
┌─────────────────────────────────────────┐
│  Artificial Intelligence (AI)          │
│  ┌───────────────────────────────────┐ │
│  │  Machine Learning (ML)            │ │
│  │  ┌─────────────────────────────┐  │ │
│  │  │  Deep Learning (DL)         │  │ │
│  │  │  Neural Networks with       │  │ │
│  │  │  multiple layers            │  │ │
│  │  └─────────────────────────────┘  │ │
│  │  Statistical methods to enable   │ │
│  │  machines to improve with data   │ │
│  └───────────────────────────────────┘ │
│  Machines performing intelligent tasks │
└─────────────────────────────────────────┘
```

- **AI**: The broadest concept - any technique enabling computers to mimic human intelligence
- **ML**: A subset of AI that uses statistical methods to enable machines to improve with experience
- **DL**: A subset of ML using neural networks with multiple layers to progressively extract higher-level features

### Real-World Examples

| Application | AI Technology | Example |
|-------------|---------------|----------|
| Healthcare | Image Recognition + ML | Disease diagnosis from X-rays |
| Transportation | Computer Vision + DL | Self-driving cars (Tesla, Waymo) |
| Finance | Pattern Recognition | Fraud detection in banking |
| E-Commerce | Recommendation Systems | Netflix, Amazon suggestions |
| Entertainment | NLP + ML | Spotify personalized playlists |
| Manufacturing | Robotics + Computer Vision | Quality control automation |

### 2. Variables and Data Types

A **variable** is a container for storing data values.

#### Basic Data Types:

| Type | Description | Example |
|------|-------------|----------|
| `int` | Integer (whole numbers) | `42`, `-10`, `0` |
| `float` | Floating-point (decimals) | `3.14`, `-0.5`, `2.0` |
| `str` | String (text) | `"Hello"`, `'Python'` |
| `bool` | Boolean (True/False) | `True`, `False` |

````python
# Integer variables
age = 25
year = 2026

# Float variables
temperature = 37.5
price = 99.99

# String variables
name = "Ahmed Ali"
city = "Lahore"
course = 'AI/ML'

# Boolean variables
is_student = True
has_passed = False

# Print all variables
print("Name:", name)
print("Age:", age)
print("City:", city)
print("Temperature:", temperature, "°C")
print("Is Student?", is_student)
````

**Key Points:**
- Variable names should be descriptive
- Use lowercase with underscores for multi-word names (`first_name`)
- Cannot start with a number
- Case-sensitive (`Name` ≠ `name`)

### 3. Type Checking and Type Casting

````python
# Check the type of a variable
x = 100
y = 3.14
z = "Python"

print("Type of x:", type(x))  # <class 'int'>
print("Type of y:", type(y))  # <class 'float'>
print("Type of z:", type(z))  # <class 'str'>
````

````python
# Type Casting - Converting between types

# String to Integer
num_str = "42"
num_int = int(num_str)
print("Converted to int:", num_int, type(num_int))

# String to Float
price_str = "99.99"
price_float = float(price_str)
print("Converted to float:", price_float, type(price_float))

# Integer to String
age = 25
age_str = str(age)
print("Converted to string:", age_str, type(age_str))

# Float to Integer (truncates decimal part)
pi = 3.14159
pi_int = int(pi)
print("Float to int:", pi_int)  # Output: 3
````

### 4. Arithmetic Operators

Python supports standard mathematical operations:

````python
# Basic arithmetic operations
a = 10
b = 3

# Addition
print("Addition: ", a + b)        # 13

# Subtraction
print("Subtraction:", a - b)      # 7

# Multiplication
print("Multiplication:", a * b)   # 30

# Division (always returns float)
print("Division:", a / b)         # 3.333...

# Floor Division (integer division)
print("Floor Division:", a // b)  # 3

# Modulus (remainder)
print("Modulus:", a % b)          # 1

# Exponentiation (power)
print("Exponent:", a ** b)        # 1000 (10^3)
````

### 5. Comparison Operators

These operators compare values and return `True` or `False`:

````python
x = 10
y = 5

# Equal to
print("x == y:", x == y)      # False

# Not equal to
print("x != y:", x != y)      # True

# Greater than
print("x > y:", x > y)        # True

# Less than
print("x < y:", x < y)        # False

# Greater than or equal to
print("x >= y:", x >= y)      # True

# Less than or equal to
print("x <= y:", x <= y)      # False
````

### 6. String Operations

Strings are sequences of characters. Python provides many ways to work with them:

````python
# Creating strings
first_name = "Ahmed"
last_name = 'Ali'
city = "Lahore"

# String concatenation (joining)
full_name = first_name + " " + last_name
print("Full Name:", full_name)

# String repetition
separator = "=" * 30
print(separator)

# String length
print("Length of name:", len(full_name))

# String methods
message = "hello world"
print("Uppercase:", message.upper())       # HELLO WORLD
print("Lowercase:", message.lower())       # hello world
print("Title Case:", message.title())      # Hello World
print("Capitalize:", message.capitalize()) # Hello world

# Replace text
text = "Python is difficult"
new_text = text.replace("difficult", "easy")
print(new_text)  # Python is easy
````

### 7. String Formatting (f-strings)

The most modern and readable way to format strings:

````python
# f-string formatting (Python 3.6+)
name = "Ahmed"
age = 25
city = "Lahore"

# Simple f-string
message = f"My name is {name} and I am {age} years old."
print(message)

# f-string with expressions
num1 = 5
num2 = 3
print(f"The sum of {num1} and {num2} is {num1 + num2}")

# f-string with formatting
pi = 3.14159265359
print(f"Pi rounded to 2 decimals: {pi:.2f}")
print(f"Pi rounded to 4 decimals: {pi:.4f}")
````

### 8. User Input

````python
# Getting input from user
# Note: input() always returns a string!

name = input("Enter your name: ")
print(f"Hello, {name}!")

# Getting numeric input (requires type casting)
age_str = input("Enter your age: ")
age = int(age_str)  # Convert string to integer
print(f"Next year you will be {age + 1} years old")
````

### 9. Comments

Comments help explain code and are ignored by Python:

````python
# This is a single-line comment
# Comments start with the hash symbol (#)

x = 10  # You can also add comments at the end of a line

"""
This is a multi-line comment.
It can span multiple lines.
Useful for longer explanations or documentation.
"""

y = 20

# Calculate the sum
total = x + y
print(f"Total: {total}")
````

Find the square root of real or complex numbers.

### Exercise 6: Area of Triangle (Task 8)

Calculate the area of a triangle using Heron's formula.

````python
# Task 8: Python Program to find the area of triangle

a = 5
b = 6
c = 7

# Uncomment below to take inputs from the user
# a = float(input('Enter first side: '))
# b = float(input('Enter second side: '))
# c = float(input('Enter third side: '))

# Calculate the semi-perimeter
s = (a + b + c) / 2

# Calculate the area using Heron's formula
area = (s * (s-a) * (s-b) * (s-c)) ** 0.5

print(f'The area of the triangle is {area:.2f}')
````

### Exercise 7: Solve Quadratic Equation (Task 9)

Solve the quadratic equation ax² + bx + c = 0.

````python
# Task 9: Solve the quadratic equation ax**2 + bx + c = 0

# Import complex math module
import cmath

a = 1
b = 5
c = 6

# Calculate the discriminant
d = (b**2) - (4*a*c)

# Find two solutions
sol1 = (-b - cmath.sqrt(d)) / (2*a)
sol2 = (-b + cmath.sqrt(d)) / (2*a)

print(f'The solutions are {sol1} and {sol2}')
````

### Exercise 8: Kilometer to Miles (Task 10)

Convert kilometers to miles.
